# Exercise 5 Big O Notation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

```

```python
#y = 0*x + 1
x = pd.Series(range(1,50))#function (inside equation)
y = 0*x + 1#equation(what we are solving for)
plt.plot(x, y, label="0(1)")#so I can copy and paste below
```

```python
#y=log*x and label curve "O(log n)"
y1 = np.log(x)
plt.plot(x, y1, label = "O(log n)")
```

```python
#y = x and label the curve "O(n)"
y2 = x
plt.plot(x, y2, label = 'O(n)')
```

```python
#y = x * log(x) and label it 'O(x log x)'
y3 = x * np.log(x)
plt.plot(x, y3, label = 'O(x log x)')
```

```python
#y = x*2 and label it 'O(x^2)'
y4 = x*x
plt.plot(x, y4, label = 'O(x^2)')
```

```python
#y = 2**x and label it 'O(2^n)'
y5 = 2**x
plt.plot(x, y5, label = 'O(2^n)')
```

```python
#y = n! and label it 'O(n!)'
y6 = x.apply(np.math.factorial) #from review....use '.apply' for strings where funtion only works for one number.
plt.plot(x, y6, label = 'O(n!)')
```

```python
# y = x*x and label it 'O(n^n)'
y7 = x*x
plt.plot(x, y7, label = 'O(n^n)')
```

```python
plt.title('Big O Notation')#Title your chart "Big O Notation"
plt.xlabel('Elements')#Label your x axis "Elements"
plt.ylabel('Operations')#Label your y axis "Operations"

plt.plot(x, y, label="0(1)")
plt.plot(x, y1, label = "O(log n)")
plt.plot(x, y2, label = 'O(n)')
plt.plot(x, y3, label = 'O(x log x)')
plt.plot(x, y3, label = 'O(x^2)')
plt.plot(x, y7, label = 'O(n^n)')

plt.legend(loc='upper left')

plt.show()
```
