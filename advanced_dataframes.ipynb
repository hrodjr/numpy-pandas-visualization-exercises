```python
#12. Create a column named mileage_difference like you did in the DataFrames exercises; this column should contain the difference between highway and city mileage for each car.
mpg = mpg.assign(mileage_difference = mpg.highway - mpg.city)
mpg.head()
```

```python
#13. Create a column named average_mileage like you did in the DataFrames exercises; this is the mean of the city and highway mileage.
mpg = mpg.assign(average_mileage=mpg.loc[:, ['city', 'highway']].mean(axis=1))
mpg.head()
```

```python
#14. Create a new column on the mpg dataset named is_automatic that holds boolean values denoting whether the car has an automatic transmission.
bools = mpg.trans.str.startswith('a')
bools
```

```python
mpg['is_auto'] = np.where(bools == True, 'auto', 'manual')
mpg.head()
```

```python
#15. Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average?
mpg.groupby('manufacturer').average_mileage.max().tail(1)
```

```python
#16. Do automatic or manual cars have better miles per gallon?
mpg.groupby('is_auto').average_mileage.max()
```

```python
menu = pd.read_sql('SELECT * FROM orders', url)
menu.head(20)
```

```python
menu.info()
```

```python
#2. What is the total price for each order?
menu['item_price'] = menu.item_price.apply(lambda value: float(value[1: ].strip()))
```

```python
menu.head(2)# I get the above error; however the "item_price" shows the '$' has been removed.
```

```python
menu.info()#with the above error, "item_price" is listed as a "float"
```

```python
#now that the "$" has been removed from "item_price" and its been converted to a float fins the total for each order_id
menu.groupby('order_id').item_price.agg('sum')#group by the order_id and sum the item prices
```

```python
#3. What are the most popular 3 items?
#Here we group by the item name and sum the quantities.
#then sort_values
menu.groupby('item_name').quantity.agg('sum').sort_values(ascending = True).tail(3)#Because I used TRUE in ascending order I took the tail (last three).
```

```python
#4. Which item has produced the most revenue?
#here we do the same except we aggregate the sum of item_price.
menu.groupby('item_name').item_price.agg('sum').sort_values(ascending = True).tail(1)
```
