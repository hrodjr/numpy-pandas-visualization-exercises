```python
import numpy as np
import pandas as pd
import math
```

```python
import matplotlib.pyplot as plt
print(plt)
```

```python
# 1. 
x = range(-50, 50)
y = [(x**2 - x + 2) for x in x]

plt.plot(x,y)

plt.title('$ y = x^2 - x +2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
# 3.
x = range(0,50)
y = [math.sqrt(n) for n in x]

plt.plot(x,y)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,50)
plt.ylim(0, 15)
```

```python
# 4.
x = range(0,50)
y = [n**3 for n in x]

plt.plot(x,y)

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
```

```python
# 5. x = range(0,50)
y = [2 ** n for n in x]

plt.plot(x,y)

plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
```
